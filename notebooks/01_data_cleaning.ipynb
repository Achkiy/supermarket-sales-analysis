## Notebook 01 - Data Cleaning
**Objetivo:** Carregar os dados brutos, identificar problemas e aplicar as correções necessárias antes de qualquer análise

**Dataset:** supermarket_raw.csv - 1000 transações de 3 filiais (Jan-Mar 2019)

```python
import pandas as pd

df = pd.read_csv('../data/supermarket_raw.csv')

print('Shape:', df.shape)
print('Colunas:', list(df.columns))
```

## Limpeza de dados

```python
# Converter Date de texto para data
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Extrair hora e converter Time para 24h
df['hour'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p').dt.hour
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p').dt.strftime('%H:%M')

# Criar colunas úteis a partir da data
df['month']       = df['Date'].dt.month
df['month_name']  = df['Date'].dt.strftime('%b')
df['day_of_week'] = df['Date'].dt.day_name()

# Remover coluna constante
df = df.drop(columns=['gross margin percentage'])

print('Feito!')
```
