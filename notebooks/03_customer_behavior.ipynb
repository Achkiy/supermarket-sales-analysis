# Supermarket Sales Analysis
## Notebook 03 - Comportamento do Cliente

**Objetivo:** Entender o perfil e comportamento de compra dos clientes - gênero, tipo de cliente de forma de pagamento

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/supermarket_clean.csv', parse_dates=['Date'])

print('Shape:' , df.shape)
```

## Receita por Gênero

```python
receita_genero = df.groupby('Gender')['Sales'].sum()

receita_genero.plot(kind='bar', color=['#EC4899', '#3B82F6'])

plt.title('Receita Total por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Receita (R$)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
print(receita_genero)
```

```python
print('Contagem por gênero:')
print(df['Gender'].value_counts())

print()
print('Ticket médio por gênero:')
print(df.groupby('Gender')['Sales'].mean().round(2))
```

## Tipo de cliente: Membro vs Normal

```python
print('Contagem por tipo:')
print(df['Customer type'].value_counts())

print()
print('Ticket médio por tipo:')
print(df.groupby('Customer type')['Sales'].mean().round(2))

print()
print('Receita total por tipo:')
print(df.groupby('Customer type')['Sales'].sum().round(2))
```

## Forma de pagamento preferida

```python
pagamento = df['Payment'].value_counts()

pagamento.plot(kind='bar', color=['#6366F1', '#F59E0B', '#10B981'])

plt.title('Transações por Forma de Pagamento')
plt.xlabel('')
plt.ylabel('Número de Transações')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(pagamento)
```
